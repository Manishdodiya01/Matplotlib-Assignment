# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in Python. It provides a wide range of functionalities for creating high-quality plots, charts, and figures.

Matplotlib is used for various purposes, including:

1. Data exploration and analysis: Matplotlib allows you to visualize your data and gain insights by creating plots such as histograms, scatter plots, and box plots.

2. Presentation and reporting: Matplotlib provides a flexible and customizable framework for creating publication-quality figures and plots, making it suitable for presentations and reports.

3. Model evaluation: It is often used to visualize model performance and evaluate the results of machine learning algorithms by creating ROC curves, precision-recall curves, and confusion matrices.

4. Trend analysis: Matplotlib can be used to create line plots to analyze trends and patterns over time or across different categories.

5. Geographic data visualization: With the help of additional toolkits like Basemap or Cartopy, Matplotlib can also handle geographical data and create maps, heatmaps, and contour plots.

Some of the plots that can be created using the Pyplot module of Matplotlib include:

1. Line plot: A basic plot that shows the relationship between two variables, typically with a continuous line connecting data points.

2. Scatter plot: A plot that represents data points as individual markers, often used to visualize the correlation or distribution of two variables.

3. Bar plot: A plot that displays categorical data with rectangular bars, where the length or height of each bar represents a specific value.

4. Histogram: A plot that shows the distribution of numerical data by dividing it into discrete bins and displaying the frequency or count of data points within each bin.

5. Pie chart: A circular plot divided into slices, where the size of each slice represents the proportion or percentage of a specific category within a dataset.

These are just a few examples, and Matplotlib offers many more types of plots and customization options to suit different visualization needs.

# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of plot that displays the relationship between two variables as a collection of individual data points. Each data point is represented by a marker on the plot, with the position of the marker determined by the values of the variables on the x and y axes.



```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.title("Scatter plot")
plt.xlabel("x-axix")
plt.ylabel("y-axis")
plt.show()
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title("Line Plot 1")

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title("Line Plot 2")

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title("Line Plot 3")


plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title("Line Plot 4")

plt.tight_layout()
plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

A bar plot, also known as a bar chart or bar graph, is a visualization that uses rectangular bars to represent categorical data. It displays the values of different categories as the heights or lengths of the bars, allowing for easy comparison between categories.



```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

print("BAR PLOT")


plt.bar(company,profit)
plt.title("Company data")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()


print("HORIZONTAL BAR PLOT")


plt.barh(company,profit)
plt.title("Company data")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
import numpy as np
import matplotlib.pyplot as plt


box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1,box2]

plt.boxplot(data , labels=['box1','box2'])
plt.title("Box plot")
plt.xlabel("Boxes")
plt.ylabel("Value")
plt.show()
```
